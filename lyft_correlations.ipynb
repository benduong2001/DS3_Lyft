```python
import pandas as pd
```

```python
rand_agents = pd.read_csv("rand_agents_table0.csv")
```

```python
# Correlation between velocity_x and velocity_y
correlation_xy = rand_agents['velocity_x'].corr(rand_agents['velocity_y'])
correlation_xy
# Negative correlation
```

```python
rand_agents.plot(x='velocity_x', y='velocity_y', kind='scatter')
```

```python
# Correlation between velocity_x and yaw
correlation_xyaw = rand_agents['velocity_x'].corr(rand_agents['yaw'])
correlation_xyaw
# Negative correlation
```

```python
rand_agents.plot(x='velocity_x', y='yaw', kind='scatter')
```

```python
# Correlation between velocity_y and yaw
correlation_yyaw = rand_agents['velocity_y'].corr(rand_agents['yaw'])
correlation_yyaw
# Positive correlation
```

```python
rand_agents.plot(x='velocity_y', y='yaw', kind='scatter')
```

```python
# Correlation between velocity_x and extent_z
correlation_xz = rand_agents['velocity_x'].corr(rand_agents['extent_z'])
correlation_xz
# No correlation
```

```python
rand_agents.plot(x='velocity_x', y='extent_z', kind='scatter')
```

```python
rand_agents.plot(x='velocity_y', y='extent_z', kind='scatter')
```
